### We want to delete the "Unnamed" column, so we have 2 approaches for that

### Approach 1: Using `.loc`
Remember that `.loc[]` is primarily label based, but may also be used with a boolean array. A boolean array of the same length as the axis being sliced, e.g. [`True, False, True]`.
Hence below we can use the boolean values instead of the column names

```python
date_data.loc[:, ~date_data.columns.str.contains("^Unnamed")] # caret(^) is added to specify column "starting with" Unnamed
```

### Approach 2: 
We can use `df.drop` and scpeicy the coumns that we want to drop by using `df.columns` and to it we pass a boolean array which stores `True` for columns we want to delete and `False` otherwise.

```python
date_data.drop(columns=date_data.columns[date_data.columns.str.contains("^Unnamed")])
```

**We can use any of the 2 approaches to drop the `Unnamed` column**

```python
date_data.drop(columns=date_data.columns[date_data.columns.str.contains("^Unnamed")], inplace=True)
```

### We missed an important function from numpy which is super useful and that's `numpy.where`
It has mutiple usages, let us look at them one at a time.

### 1. We want to return index when a particular condition is met

```python
arr = np.random.randint(10, 50, 10)
arr
```

```python
np.where(arr > 20)
```

```python
arr[arr > 20] # this can also be done using np.where
```

```python
arr[np.where(arr > 20)]
```

### 2. We want to modify the array when a condition is met

Suppose we want to modify the value of array to -1 when value is less than 20 and keep it unchanged otherwise

```python
np.where(arr < 20, -1, arr)
```

### 3. We want to replace original value of array when a condition is met and subsititue a different value it condition is not met.

Suppose we have an array of age and we want to subsitute `teen` when age is between 13 to 19 and `not teen` otherwise 

```python
school_data = pd.DataFrame({'Age': np.random.randint(5, 19, 20)})
```

```python
school_data.head()
```

We create a new column called tag and assign `teen` and `not teen` value, depending on the age

```python
9 in range(1, 5)
```

```python
school_data['Tag'] = np.where((school_data["Age"] >= 13) & (school_data["Age"] <= 19), "teen", "not teen")
school_data
```

# Multi indexing in Pandas

```python
stocks_data = pd.read_csv("http://bit.ly/smallstocks")
```

```python
stocks_data
```

```python
# This is a regular index
stocks_data.index
```

```python
stocks_data.groupby("Symbol")["Close"].mean()
```

```python
ser = stocks_data.groupby(["Symbol", "Date"])["Close"].mean()
ser
```

```python
type(ser)
```

```python
ser.index # So ser is series with a multi-index
```

### A multi-index adds an extra dimension to our data, a `Series` with one/normal index is one dimensional but a Series with multi-index is kind of similair to 2 dimensional. 

### A `dataframe` with normal index is 2 dimensional but with a multi-index is atleast 3 dimensional.

**We can unstack a Series with multi-index and it will get converted to a Dataframe as Series with a multi-index is similar to 2 dimensional object**

```python
ser.unstack()
```

#### Above structure can also be acheived by making a `pivot_table`

```python
stocks_data.pivot_table(index='Symbol', columns='Date', values='Close')
```

### Unstack - It would move/rotate the inner most row index and would make it a column index
### Stack - It would move/rotate the inner most column index and would make it a row index

## Selecting data from a Multi-index Series

### 1. To select data using the outer most index, do `ser.loc['AAPL']` <br>
### 2. To select data from the outer and inner most index, do `ser.loc['AAPL', '2016-10-03']` <br>
### 3. To select data just from outer index, use `ser.loc[:, '2016-10-03']`

```python
ser
```

```python
ser.loc['AAPL']
```

```python
ser.loc['AAPL', '2016-10-03']
```

```python
ser.loc[:, '2016-10-03']
```

```python
ser.loc[['AAPL', 'CSCO'], ['2016-10-03', '2016-10-04']]
```

### Looking at the above example - we understand that series with multi-index is similar to a dataframe with normal index because we use `df.loc[[rows list], [columns list]]` to pull data from dataframe and we do same kind of thing for Series with Multi-index where the outer most index becomes the row and the inner most index becomes the column

```python
df = ser.unstack()
df
```

```python
df.index
```

```python
df.loc[['AAPL', 'CSCO'], ['2016-10-03', '2016-10-04']]
```

```python
stocks_data.loc[('AAPL', '2016-10-03'), :]
```

**Display data from mutiple stocks** - We again use `[ ]` and pass it as a list

```python
stocks_data.loc[(['AAPL', 'MSFT'], '2016-10-03'), :]
```

```python
stocks_data.loc[('AAPL', ['2016-10-03', '2016-10-04']), :]
```
